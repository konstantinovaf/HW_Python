Задание 1

Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

```python
def rating_classificator(row):
    param = row['rating']
    """This function classificate film rating"""
    if param <=2:
        return "низкий рейтинг"
    elif 4 >= param > 2:
        return "средний рейтинг"
    elif param == 4.5 or param == 5: 
        return "высокий рейтинг"
    else:
        return "Рейтинг за границами оценки. Проверьте данные"
    
```

```python
import pandas as pd
ratings = pd.read_csv('ratings.csv')
ratings.head()
# stats = pd.read_excel('ratings.xls')
# stats.columns = ['group', 'phrase', 'effect', 'ad_id', 'title', 'text', 'link']
# stats.head()
```

```python
ratings['class'] = ratings.apply(rating_classificator, axis=1)
ratings.head(15)
```
